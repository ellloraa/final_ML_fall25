```python
import json
import numpy as np

def load_posts(filename):
    with open(filename, 'r') as f:
        return [json.loads(line) for line in f]

# Placeholder sentiment function (replace with real model)
def get_sentiment_score(text):
    return 0.0  # stub

def is_controversial(post, avg_comment_count, threshold=0.5):
    upvotes = post.get("vote_total", 0)
    comments = post.get("comment_count", 0)
    text = post.get("text", "")
    
    # Feature 1: comment ratio
    comment_ratio = comments / (avg_comment_count + 1)
    
    # Feature 2: engagement ratio
    engagement_ratio = comments / (upvotes + 1)
    
    # Feature 3: sentiment polarity (stubbed)
    sentiment = get_sentiment_score(text)
    sentiment_variance = abs(sentiment)  # placeholder
    
    # Weighted score
    score = (
        0.4 * comment_ratio +
        0.4 * engagement_ratio +
        0.2 * sentiment_variance
    )
    
    # Binary label
    return 1 if score >= threshold else 0

# Run model
def run_model(filename):
    posts = load_posts(filename)
    avg_comments = np.mean([p.get("comment_count", 0) for p in posts])
    
    labeled_posts = []
    for post in posts:
        label = is_controversial(post, avg_comments)
        labeled_posts.append({
            "id": post["id"],
            "text": post["text"],
            "controversial": label
        })
    
    return labeled_posts

# Example usage
if __name__ == "__main__":
    results = run_model("posts.jsonl")
    controversial_posts = [r for r in results[:150] if r["controversial"] == 1]
    for r in controversial_posts:
        print(r)
```
