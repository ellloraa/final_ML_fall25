```python
import json

file_paths = ["reddit_yikyak_dataset.json", "reddit_yikyak_riana.json"]
merged_data = []

# Read data from both files and append to a list
for file_path in file_paths:
    with open(file_path, 'r') as infile:
        data = json.load(infile)
        # Assuming each file contains a list of records (common format)
        if isinstance(data, list):
            merged_data.extend(data)
        else:
            # Handle if files contain single objects (e.g. merge object keys)
            merged_data.append(data) 

# Write the combined data to a new file
with open("merged_file.json", "w") as outfile:
    json.dump(merged_data, outfile, indent=4)
```
